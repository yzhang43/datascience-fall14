```python
####Construct a dataframe with two columns: country_name, num_players. Use groupby.###
```

```python
Inner_join = pd.merge(countries, players, on = 'country_id', how = 'inner');
Inner_join
result_3 = pd.DataFrame({'num_players' : Inner_join.groupby('name_x').size()}).reset_index()
result_3.columns = ['country_name','num_players']
result_3
```

```python
####Convert the above left outer join query (last question in the SQL assignment) into pandas equivalent.####
```

```python
player_1975 = players[players['birthdate'].str.contains('1975')];
left_out_join_res = pd.merge(countries, player_1975, on = 'country_id', how = 'left');
result_4 = pd.DataFrame({'num_of_players_1975' : left_out_join_res.groupby('name_x').birthdate.count()}).reset_index()
result_4.columns = ['country_name','num_of_players_1975']
result_4
```
